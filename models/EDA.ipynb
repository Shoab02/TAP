```python
#Total house sold year wise
yearly_sold = df.groupby(['YrSold'])['YrSold'].count()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
yearly_sold.plot(kind='bar', cmap='Pastel1')

plt.subplot(1,2,2)
plt.plot(yearly_sold)
```

<h3>Year 2016 was the most successful in house sales. <br>

We can notice a pattern for every four years starting 2008.</h3>


```python
#Total house sold hallway type wise
yearly_sold = df.groupby(['HallwayType'])['HallwayType'].count()

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
yearly_sold.plot(kind='bar', cmap="Pastel2")

plt.subplot(1,2,2)
plt.plot(yearly_sold)
```

<h3>Clearly terraced hallway is the most preferred</h3>

```python
fig, axes = plt.subplots(2, 1, figsize=(18, 10), sharey=False)
fig.suptitle('Floor# Analysis')

floor_count = df['Floor'].value_counts()

sns.barplot(x=floor_count.index, y=floor_count.values, alpha=0.9, ax=axes[0])
axes[0].set_xlabel('Floor',fontsize=12)
axes[0].set_ylabel('Houses sold',fontsize=12)
axes[0].set_title('Houses sold based on Floor',fontsize=14)

sns.lineplot(x='Floor', y='SalePrice',hue = 'HallwayType', data=df, ax=axes[1])
axes[1].set_title('Floor# Vs Sales price',fontsize=14)

plt.show()
```

<h3>Majority of the houses sold were for floor range 1-22. <br> A major factor affecting this is the cost which can be visulized from the line plot just above</h3>

```python
bin_labels = ['0-600', '600-900', '900-1200', '1200-1500','1500-2000','2000+']
bin_cuts = [0, 600, 900, 1200, 1500,2000,2500]
df['sqft cat'] = pd.cut(df['Size(sqf)'], bins=bin_cuts, labels=bin_labels)
```

```python
sqft_sold = df['sqft cat'].value_counts()

plt.figure(figsize=(8,5))

sns.barplot(x=sqft_sold.index, y=sqft_sold.values, alpha=0.9)
plt.xlabel('sqft',fontsize=18)
plt.ylabel('Houses sold',fontsize=18)
plt.title('Houses sold based on sqft',fontsize=18)
```

<h3>It's better to visualize the sqft info when grouped into bins and calculating the houses sold for each group. <br>
As we can see, most houses sold are between the 900-1200 sqft range</h3>

```python

plt.figure(figsize=(10,5))
sns.lineplot(x='sqft cat', y='SalePrice', data=df)
plt.title('Size(sqf) Vs Sales price',fontsize=18)
```

<h3>The sales price is linearly related to sqft. <br>
The anamoly at the end the line could be due to less data for the bin above 2000+</h3>
